## House pricing problem

```python
## imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Load the data
house_sizes = np.array([750, 800, 850, 900, 950, 1000, 1100, 1200, 1300, 1400])
house_prices = np.array([150, 180, 200, 210, 250, 275, 300, 320, 340, 360])
```

```python
## find the mean of the data
X_mean = np.mean(house_sizes)
Y_mean = np.mean(house_prices)
```

```python
## compute slope, intercept using least square method
numerator = np.sum((house_sizes - X_mean) * (house_prices - Y_mean))
denominator = np.sum((house_sizes - X_mean) ** 2)
m = numerator / denominator
b = Y_mean - (m * X_mean)
```

```python
## generate best fit line
Y_pred = m * house_sizes + b
```

```python
## predict price for 115 sq ft
predicted_price = m * 115 + b
print("Predicted price for 115 sq ft is: ", predicted_price)
```

```python
## plot the data
plt.figure(figsize=(8, 5))
plt.scatter(house_sizes, house_prices, color='red', label='Data points')
plt.plot(house_sizes, Y_pred, color='blue', label='Best fit line')
plt.xlabel('House Size (sq ft)')
plt.ylabel('House Price ($k)')
plt.title('House Price Prediction')
plt.legend()
plt.grid()
plt.show()
```
